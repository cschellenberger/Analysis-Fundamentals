```python
import pandas as pd
```

```python
attendance_data.plot(x='Days Absent', 
                     y='Final Grade', 
                     kind='scatter', 
                     title='Relationship between absences and final grades')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.scatter(attendance_data['Days Absent'], attendance_data['Final Grade'], label='data', color='red', marker='o')

sns.regplot(x='Days Absent', y='Final Grade', data=attendance_data, color='blue', scatter=None, label='order 1')

sns.regplot(x='Days Absent', y='Final Grade', data=attendance_data, color='green', scatter=None, label='order 2', order=2)

plt.legend(loc='upper right')
plt.ylim(0, 120)
plt.show()
```

```python
sns.residplot(x='Days Absent', y='Final Grade', data=attendance_data, color='green')
```

```python
sns.residplot(x='Days Absent', y='Final Grade', data=attendance_data, color='green', order=2)
```

```python
sns.jointplot(x='Days Absent', y='Final Grade', data=attendance_data)
```
